```python
#plot the variation in the gas prices of USA over the year

plt.figure(figsize=(5,3), dpi=200)


plt.xticks(gas['Year'], rotation = 90)
plt.plot(gas['Year'], gas['USA'])
plt.xlabel('Year')
plt.ylabel('Gas prices of USA')


plt.show()
```

```python
#plot the variation in the gas prices of USA v/s over the year

plt.figure(figsize=(8,5), dpi=200)


plt.xticks(gas['Year'], rotation = 60)

plt.plot(gas['Year'], gas['USA'], 'r.-',label = 'USA Gas Price over the years')
plt.plot(gas['Year'], gas['Canada'], 'g.-',label = 'Canada Gas Price over the years')
plt.title('Price comparison of USA v/s Canadian gas prices')
plt.xlabel('Year', fontdict = {'fontsize':'14'}) 
plt.ylabel('Price is US $',fontdict = {'fontsize':'14'})

plt.legend()

plt.show()
```

```python
#Plot the variation of gas prices for all the countries


plt.figure(figsize=(20,12), dpi=800)
plt.xticks(gas['Year'], rotation = 20)


for country in gas:
    if(country != 'Year'):
        plt.plot(gas['Year'], gas[country],marker = '*',lw=3.25, label = 'Price variation of prices for ' + str(country))
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=5)
plt.xlabel('Year', fontdict = {'fontsize' : '10'})
plt.ylabel('Prices in US dollars', fontdict = {'fontsize' : '20'})
plt.title('Variation of gas prices for all the countries',fontdict = {'fontsize' : '35'})
plt.show()
```
